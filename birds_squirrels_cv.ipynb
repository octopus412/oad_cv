```python
def rename(path):
  for i, folder in enumerate(os.listdir(path)):
    for j, file in enumerate(os.listdir(path + folder)):
      os.rename(os.path.join(path + folder, file), os.path.join(path + folder, folder + file))
```

```python
def resizeAllImagesInDir(pathToDir):
    import os
    import cv2

    for filename in os.listdir(pathToDir):
        abs_filename = os.path.join(pathToDir, filename)
        if os.path.isfile(abs_filename):
            # Process the file here
            img = cv2.imread(abs_filename)

            if img is not None:
              img = cv2.resize(img, (224, 224))
              cv2.imwrite(abs_filename, img)
            else:
              os.remove(abs_filename)
```

```python
resizeAllImagesInDir('Images/train/0squirrels')
resizeAllImagesInDir('Images/train/1birds')
resizeAllImagesInDir('Images/train/2trash')
resizeAllImagesInDir('Images/val/0squirrels')
resizeAllImagesInDir('Images/val/1birds')
resizeAllImagesInDir('Images/val/2trash')
```

```python
contours, h = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
frameCopy = frame.copy()
cv2.drawContours(frameCopy, contours, -1, (255, 0, 0), 2)
plt.imshow(frameCopy)
```

```python
def drawRect(contours, frame):
    rectangles = []
    res_rect = []

    for i in range(0, len(contours)):
        x, y, w, h = cv2.boundingRect(contours[i])
        rectangles.append([x, y, x + w, y + h])
    rectangles.sort()

# фильтрация
    for i in range(len(rectangles)):
        if not rectangles[i]:
            continue

        x, y, X, Y = rectangles[i]

        if (X - x <= 30) | (Y - y <= 30):
            rectangles[i] = []
            continue

        for j in range(len(rectangles)):
            if (i == j) | (not rectangles[j]):
                continue
            x_j, y_j, X_j, Y_j = rectangles[j]

            if (x_j <= x) & (X_j >= X) & (y_j <= y) & (Y_j >= Y):
                rectangles[i] = []
                break
            if (x <= x_j) & (X >= X_j) & (y <= y_j) & (Y >= Y_j):
                rectangles[j] = []
                continue

    for i in range(len(rectangles)):
        if rectangles[i]:
            cv2.rectangle(frame, (rectangles[i][0], rectangles[i][1]), (rectangles[i][2], rectangles[i][3]), (255, 215, 0), 2)

    return rectangles
```

```python
def changeFrame(rectangles, frame):
    bright_matr = np.ones(frame.shape, dtype = "uint8") * 50
#     higher_contrast_matr = np.ones(frame.shape, dtype = "uint8") * 2.

    brighter_frame = cv2.add(frame, bright_matr)
#     higher_frame = np.uint8(cv2.multiply(np.float64(brighter_frame), higher_contrast_matr))

    for i in range(len(rectangles)):
        if rectangles[i]:
            x, y, X, Y = rectangles[i]
#             frame[y:Y, x:X] = higher_frame[y:Y, x:X]
            frame[y:Y, x:X] = brighter_frame[y:Y, x:X]

            if (X - x < 100):
              cv2.rectangle(frame, (x - 3, y - 50), (x + 100, y), (255, 215, 0), -2)
            else:
              cv2.rectangle(frame, (x - 3, y - 50), (X, y), (255, 215, 0), -2)

            cv2.putText(frame, "squirrel: 0.000", (x, y - 35), cv2.FONT_HERSHEY_PLAIN, 0.8, (0, 0, 0), thickness =1, lineType = cv2.LINE_AA)
            cv2.putText(frame, "Image", (x, y - 20), cv2.FONT_HERSHEY_PLAIN, 1., (0, 0, 0), thickness =1, lineType = cv2.LINE_AA)
            cv2.putText(frame, "Image", (x, y - 5), cv2.FONT_HERSHEY_PLAIN, 1., (0, 0, 0), thickness =1, lineType = cv2.LINE_AA)
```

```python
frame1 = frame.copy()
rect = drawRect(contours, frame1)
bird = changeFrame(rect, frame1)
plt.figure(figsize=(10, 12))
plt.imshow(frame1)
# plt.imshow(bird)
```
